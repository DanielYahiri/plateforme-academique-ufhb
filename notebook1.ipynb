Introduction des Enregistrements en python


```python
from dataclasses import dataclass
#exemple d'enregistrement livre
@dataclass
class Livre:
    titre: str     #on precise le type de chaque input
    auteur: str
    pages: int

```

Liste pour stocker nos instances

```python
bibliotheque = []
nombre = int(input("Combien de livres voulez-vous enregistrer ? "))
```

Recuperation des instances


```python
for i in range(nombre):
    print(f"\n Livre n°{i+1} ")
    t = input("Titre : ")
    a = input("Auteur : ")
    p = int(input("Pages : "))
    nouveau_livre = Livre(titre=t, auteur=a, pages=p)
    bibliotheque.append(nouveau_livre)
```

Affichage de la collection

```python

print(f"\nContenu de la bibliothèque :")
for livre in bibliotheque:
    print(f"\n titre : {livre.titre} ,auteur :{livre.auteur},pages:,{livre.pages} ")
```

```python
def bissextile(annee):
    if (annee%4==0 and annee%100!=0) or (annee%400==0)  :
        return 'bissextile'
    else :
        return 'pas bissextile'


```
